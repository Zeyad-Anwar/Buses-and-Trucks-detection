```python
from ultralytics import YOLO
import cv2
import glob
import os
import matplotlib.pyplot as plt
```

```python
pictures_path = "/content/pictures"
os.makedirs(pictures_path, exist_ok=True)
```

```python
video_path = "/content/video"
os.makedirs(video_path, exist_ok=True)
video_path = video_path + "/test_video.mp4"
```

```python
model_path = "/content/model"
os.makedirs(model_path, exist_ok=True)
```

```python
model_path = model_path + "/my_model.pt"
```

```python
model = YOLO(model_path)
```

```python
image_paths = glob.glob(pictures_path + "/*.jpg")
image_paths = image_paths + glob.glob(pictures_path + "/*.jpeg")
image_paths
```

```python
for img_path in image_paths:
    results = model.predict(img_path, save=False)

    annotated_img = results[0].plot()
    annotated_img = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(8, 6))
    plt.imshow(annotated_img)
    plt.axis("off")
    plt.show()
```

```python
results = model(video_path, save=True, project='/content/video', name='result_video')
```
